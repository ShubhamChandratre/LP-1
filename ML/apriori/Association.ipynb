```python
df.dropna(how="all")
```

```python
records=[]
for i in range(0,7500):
    records.append([str(df.values[i,j]) for j in range(0,20)]) 
```

```python
association_rules=apriori(records,min_support=0.001,min_confidence=0.7,min_lift=1.2,min_length=2)
association_results=list(association_rules)
```
